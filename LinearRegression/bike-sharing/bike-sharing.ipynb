### Verifica se existem dados nulos no dataset

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
lYText = ['Primavera',u'Vero','Outono', 'Inverno']
lXText = ['Domingo', 'Segunda', u'Tera', 'Quarta', 'Quinta', 'Sexta', u'Sbado']
f, ax  = plt.subplots(figsize=(9, 6))
data   = df.pivot_table("registered", "weekday", "season")
ax     = sns.heatmap(data, linewidths=.5, ax=ax , cmap='Blues')

ax.set_xlabel(u'Estaes do ano')
ax.set_ylabel('Dias da semana')
ax.set_yticklabels(lXText, fontsize=10, rotation=30)
ax.set_xticklabels(lYText, fontsize=10, rotation=30)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
y = df[[u'temp']]
X = df[[ 'registered', 'casual', 'atemp']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
lm = LinearRegression()

lm.fit(X_train, y_train)
```

```python
print(lm.intercept_)
```

```python
print(lm.coef_)
```

```python
coefs = pd.DataFrame(lm.coef_.reshape(-1), X.columns, columns=['Coeficientes'])
```
